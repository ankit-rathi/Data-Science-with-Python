```python

def plot_stock_buy_sell(stock_symbol):
  # Don't forget to import pandas_datareader exactly in this way
  import yfinance as yfin
  # Then replace this in for loop
  stock_df = yfin.Ticker(stock_symbol).history(period='2y',interval='1d')[map(str.title, ['open', 'close', 'low', 'high', 'volume'])]
  stock_df['50DMA'] = round(stock_df['Close'].rolling(window = 50, min_periods = 1).mean(),0)
  stock_df['150DMA'] = round(stock_df['Close'].rolling(window = 150, min_periods = 1).mean(),0)

  import datetime
  end_date = datetime.date.today().strftime('%Y-%m-%d')
  start_date = (datetime.date.today() + datetime.timedelta(days=-500)).strftime('%Y-%m-%d')

  stock_df = stock_df[(stock_df.index <= end_date) & (stock_df.index >= start_date)]
  stock_df = stock_df.reset_index()
  stock_df['Date'] = stock_df['Date'].dt.date
  stock_df = stock_df.set_index('Date')

  import matplotlib.pyplot as plt
  # plot close price, short-term and long-term moving averages
  plt.figure(figsize = (25,10))
  plt.tick_params(axis = 'both', labelsize = 10)
  stock_df['Close'].plot(color = 'k', lw = 1, label = 'Close')  
  stock_df['150DMA'].plot(color = 'm', lw = 1, label = '150DMA')
  stock_df['50DMA'].plot(color = 'b', lw = 1, label = '50DMA')
  #stock_df['Volume'].plot(color = 'k', lw = 1, label = 'Volume')

  plt.ylabel('Price in ₹', fontsize = 10 )
  plt.xlabel('Date', fontsize = 10 )
  plt.title(stock_symbol + '  : Buy/Sell Signal - Daily', fontsize = 20)
  plt.legend()
  plt.grid()
  plt.show()

  # plot close price, short-term and long-term moving averages
  plt.figure(figsize = (25,3))
  plt.tick_params(axis = 'both', labelsize = 14)
  stock_df['Volume'].plot.bar(color = 'k', label = 'Volume')
  plt.ylabel('Volume', fontsize = 14 )
  plt.grid()
  plt.show()
```

```python
plot_stock_buy_sell('^NSEI')
```

```python
portfolio = ['alkylamine','asianpaint','axisbank','bergepaint','bhartiartl','canbk','dmart','federalbnk','havells','hdfcbank',
             'hdfclife','hindunilvr','ibrealest','icicibank','indusindbk','infy','irctc','itc','mhril','motilalofs',
             'nationalum','ntpc','paras','reliance','sail','tataconsum','tatamotors','tatapower','tatasteel','tcs','zeel']

nextbuy = ['sbin','titan','ultracemco','kotakbank']

watchlist = ['equitasbnk','hcltech','hdfcamc','heromotoco','iex','infobean','jublfood','lti','tipsindltd',
             'mastek','mindtree','relaxo']

skipped = ['canbk', 'pageind','hdfcamc','equitasbnk','irctc', 'relaxo', 'tipsindltd']

stock_ids = list(set(portfolio + nextbuy + watchlist) - set(skipped))
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

name_list = []
cp_list = []
pc_list = []
th_list = []
tl_list = []
yh_list = []
yl_list = []

overall_rating_lst = []
management_rating_lst = []
valuation_rating_lst = []
efficiency_rating_lst = []
financials_rating_lst = []
essentials_lst = []

for share_name in stock_ids:
    
    print(share_name)
    link_blueprint = "https://ticker.finology.in/company/"

    page = requests.get(link_blueprint + share_name.upper()  + "?mode=C")
    link_soup = BeautifulSoup(page.content, 'html.parser')

    # Name
    name = link_soup.find(id="mainContent_ltrlCompName").get_text()

    name_list.append(share_name.upper() + '.NS')

    # Price
    price_class = link_soup.find(class_="d-block h1 currprice")
    current_price = price_class.find(class_="Number").get_text()

    cp_list.append(current_price)

    # Price change
    price_change = link_soup.find(id="mainContent_pnlPriceChange").get_text()

    pc_list.append(price_change)

    # Today's High and Low
    today_high = link_soup.find(id="mainContent_ltrlTodayHigh").get_text()
    today_low = link_soup.find(id="mainContent_ltrlTodayLow").get_text()

    th_list.append(today_high)
    tl_list.append(today_low)

    # 52 week high and low
    year_high = link_soup.find(id="mainContent_ltrl52WH").get_text()
    year_low = link_soup.find(id="mainContent_ltrl52WL").get_text()

    yh_list.append(year_high)
    yl_list.append(year_low)
    
    
    overall_rating = link_soup.find(id="mainContent_ltrlOverAllRating")['style'].split(":",1)[1].split(";",1)[0] 
    management_rating = link_soup.find(id="mainContent_ManagementRating")['style'].split(":",1)[1].split(";",1)[0] 
    valuation_rating = link_soup.find(id="mainContent_ValuationRating")['style'].split(":",1)[1].split(";",1)[0] 
    efficiency_rating = link_soup.find(id="mainContent_EfficiencyRating")['style'].split(":",1)[1].split(";",1)[0] 
    financials_rating = link_soup.find(id="mainContent_FinancialsRating")['style'].split(":",1)[1].split(";",1)[0] 
    
    overall_rating_lst.append(overall_rating)
    management_rating_lst.append(management_rating)
    valuation_rating_lst.append(valuation_rating)
    efficiency_rating_lst.append(efficiency_rating)
    financials_rating_lst.append(financials_rating)
    
    test_list = link_soup.find(id="mainContent_updAddRatios").text.replace("  ", "").split('\n')

    while("" in test_list) :
        test_list.remove("")
    while("\r" in test_list) :
        test_list.remove("\r")        
    while(" \r" in test_list) :
        test_list.remove(" \r")
    while("₹\xa0\r" in test_list) :
        test_list.remove("₹\xa0\r")
    while("Add Your Ratio" in test_list) :
        test_list.remove("Add Your Ratio")
        
    essentials_lst.append(test_list)


df_portfolio_scrap = pd.DataFrame(
    {
        'Stock': name_list,
        'Overall Rating': overall_rating_lst,
        'Management Rating': management_rating_lst,
        'Valuation Rating': valuation_rating_lst,
        'Efficiency Rating': efficiency_rating_lst,
        'Financials Rating': financials_rating_lst,
        'Current Price': cp_list,
        'Year High': yh_list,
        'Year Low': yl_list,
        'Company Essentials': essentials_lst

    })

df_portfolio_scrap
```

```python
df_essentials = pd.DataFrame()

for elements in df_portfolio_scrap['Company Essentials']:
    
    def Convert(a):
        it = iter(a)
        res_dct = dict(zip(it, it))
        return res_dct
    print(Convert(elements))

    tmp = pd.DataFrame([Convert(elements)], columns=Convert(elements).keys())
    df_essentials = pd.concat([df_essentials,tmp])    

df_essentials.reset_index(drop=True,inplace=True)    
df_portfolio =  pd.concat([df_portfolio_scrap, df_essentials], axis=1, ignore_index=True)
df_portfolio.columns = df_portfolio_scrap.columns.append(df_essentials.columns)
```

```python
df_portfolio['Overall Rating'] = df_portfolio['Overall Rating'].astype(float).fillna(0.0)
df_portfolio['Management Rating'] = df_portfolio['Management Rating'].astype(float).fillna(0.0)
df_portfolio['Valuation Rating'] = df_portfolio['Valuation Rating'].astype(float).fillna(0.0)
df_portfolio['Efficiency Rating'] = df_portfolio['Efficiency Rating'].astype(float).fillna(0.0)
df_portfolio['Financials Rating'] = df_portfolio['Financials Rating'].astype(float).fillna(0.0)

df_portfolio['Weighted Rating'] = (df_portfolio['Overall Rating']*5) + (df_portfolio['Management Rating']*4) + \
                                    (df_portfolio['Valuation Rating']*3) + (df_portfolio['Financials Rating']*2) + \
                                        df_portfolio['Efficiency Rating']

df_portfolio['Market Cap'] = df_portfolio['Market Cap'].str.replace('₹ ', '').str.replace(' Cr.\r', '')
df_portfolio['Market Cap'] = df_portfolio['Market Cap'].astype(float).fillna(0.0)
df_portfolio = df_portfolio.rename(columns={'Market Cap': 'Market Cap (₹ Cr.)'})

df_portfolio['Enterprise Value'] = df_portfolio['Enterprise Value'].str.replace('₹ ', '').str.replace(' Cr.', '')
df_portfolio['Enterprise Value'] = df_portfolio['Enterprise Value'].astype(float).fillna(0.0)
df_portfolio = df_portfolio.rename(columns={'Enterprise Value': 'Enterprise Value (₹ Cr.)'})

df_portfolio['No. of Shares'] = df_portfolio['No. of Shares'].str.replace(' Cr.\r', '')
df_portfolio['No. of Shares'] = df_portfolio['No. of Shares'].astype(float).fillna(0.0)
df_portfolio = df_portfolio.rename(columns={'No. of Shares': 'No. of Shares (Cr.)'})

df_portfolio['P/E'] = df_portfolio['P/E'].str.replace('\r', '')
df_portfolio['P/E'] = df_portfolio['P/E'].astype(float).fillna(0.0)

df_portfolio['P/B'] = df_portfolio['P/B'].str.replace('\r', '')
df_portfolio['P/B'] = df_portfolio['P/B'].astype(float).fillna(0.0)

df_portfolio['ROE \r'] = df_portfolio['ROE \r'].str.replace('%\r', '')
df_portfolio['ROE \r'] = df_portfolio['ROE \r'].astype(float).fillna(0.0)
df_portfolio = df_portfolio.rename(columns={'ROE \r': 'ROE (%)'})

df_portfolio['ROCE'] = df_portfolio['ROCE'].str.replace('%\r', '')
df_portfolio['ROCE'] = df_portfolio['ROCE'].astype(float).fillna(0.0)
df_portfolio = df_portfolio.rename(columns={'ROCE': 'ROCE (%)'})

df_portfolio['Profit Growth\r'] = df_portfolio['Profit Growth\r'].str.replace('%\r', '')
df_portfolio['Profit Growth\r'] = df_portfolio['Profit Growth\r'].astype(float).fillna(0.0)
df_portfolio = df_portfolio.rename(columns={'Profit Growth\r': 'Profit Growth (%)'})

df_portfolio['Sales Growth'] = df_portfolio['Sales Growth'].str.replace('%\r', '')
df_portfolio['Sales Growth'] = df_portfolio['Sales Growth'].astype(float).fillna(0.0)
df_portfolio = df_portfolio.rename(columns={'Sales Growth': 'Sales Growth (%)'})

df_portfolio['Div. Yield'] = df_portfolio['Div. Yield'].str.replace('%\r', '')
df_portfolio['Div. Yield'] = df_portfolio['Div. Yield'].astype(float).fillna(0.0)
df_portfolio = df_portfolio.rename(columns={'Div. Yield': 'Div. Yield (%)'})
```

```python
df_portfolio = df_portfolio.sort_values(by = 'Weighted Rating', ascending = False).reset_index(drop=True)
```

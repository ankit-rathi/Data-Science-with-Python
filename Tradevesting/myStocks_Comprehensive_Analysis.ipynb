```python
cols_a = ['Stock','Dev%_200','Upside%ATH','Criteria','Strategy','StarStock']
cols_b = ['Stock','Current P/L%','Weightage%','Investment']
tmp_a = df_ps_features[cols_a]
tmp_b = df_pf_analysis[cols_b]
tmp_df = pd.merge(tmp_a, tmp_b, on='Stock', how='inner')
tmp_df.sort_values(by = 'Dev%_200', ascending=True)
```
